```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_one = "2017-18_teamBoxScore.csv"
```

```python
df = pd.read_csv(file_one, encoding="ISO-8859-1")
```

```python
df.head()
```

```python
list(df)
```

```python
df['gmDate'] =  pd.to_datetime(df['gmDate'], format='%Y-%m-%d')
df = df.set_index(df['gmDate'])
```

```python
df_bydate = df.groupby(pd.Grouper(freq="M"))
g = df_bydate.mean()
```

```python
g['teamFG%'].head(7)
```

```python
g['teamFG%'].plot()
plt.savefig("2017-18monthlyFG%")
```
